**STORY GNERATION MODEL USING GPT2**

```python
from transformers import pipeline

# Load the basic GPT-2 model
generator = pipeline("text-generation", model="gpt2")

prompt = "In a forest where trees could talk,"
story = generator(prompt, max_length=150, temperature=0.9)

print(story[0]["generated_text"])
```

**STORY GENERATION  MODEL USING GPT2-MEDIUM MODEL**

```python
from transformers import pipeline

story_gen = pipeline("text-generation", model="gpt2-medium")

prompt = "In the heart of a forgotten realm, a young warrior set out on a journey."
story = story_gen(prompt, max_length=200, temperature=0.8, do_sample=True)

print(story[0]['generated_text'])
```
